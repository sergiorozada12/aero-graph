```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
DATA_PATH = "data/results1703/"
```

```python
def get_data(results):
    unbalanced_acc = []
    balanced_acc = []
    balanced_zeros_acc = []

    unbalanced_prec = []
    balanced_prec = []
    balanced_zeros_prec = []

    unbalanced_rec = []
    balanced_rec = []
    balanced_zeros_rec = []
    
    unbalanced_f1 = []
    balanced_f1 = []
    balanced_zeros_f1 = []
    
    for node in results.keys():
        unbalanced_acc.append(results[node]['unbal']['acc'])
        balanced_acc.append(results[node]['bal']['acc'])
        balanced_zeros_acc.append(results[node]['bal_zeros']['acc'])

        unbalanced_prec.append(results[node]['unbal']['prec'])
        balanced_prec.append(results[node]['bal']['prec'])
        balanced_zeros_prec.append(results[node]['bal_zeros']['prec'])

        unbalanced_rec.append(results[node]['unbal']['rec'])
        balanced_rec.append(results[node]['bal']['rec'])
        balanced_zeros_rec.append(results[node]['bal_zeros']['rec'])
        
        unbalanced_f1.append(results[node]['unbal']['f1'])
        balanced_f1.append(results[node]['bal']['f1'])
        balanced_zeros_f1.append(results[node]['bal_zeros']['f1'])
    
    return {'unbalanced_acc': unbalanced_acc,
            'unbalanced_prec': unbalanced_prec,
            'unbalanced_rec': unbalanced_rec,
            'unbalanced_f1': unbalanced_f1,
            'balanced_acc': balanced_acc,
            'balanced_prec': balanced_prec,
            'balanced_rec': balanced_rec,
            'balanced_f1': balanced_f1,
            'balanced_zeros_acc': balanced_zeros_acc,
            'balanced_zeros_prec': balanced_zeros_prec,
            'balanced_zeros_rec': balanced_zeros_rec,
            'balanced_zeros_f1': balanced_zeros_f1}
```

```python
def plot_hist_features(metrics_1, metrics_2, exp_type):
    plt.figure(figsize=[16, 3])
    plt.hist(metrics_1[exp_type], range=[0, 1], bins=50, color='b', alpha=.5)
    plt.hist(metrics_2[exp_type], range=[0, 1], bins=50, color='r', alpha=.5)
    plt.xlabel(exp_type)
    plt.legend(['LR', 'RF'])
    plt.show()
```

# 1) Decision Tree

```python
with open(DATA_PATH + 'results_DT_lr_node.json', 'r') as f:
    results_DT_lr = json.load(f)
    
with open(DATA_PATH + 'results_DT_rf_node.json', 'r') as f:
    results_DT_rf = json.load(f)
```

```python
metrics_lr = get_data(results_DT_lr)
metrics_rf = get_data(results_DT_rf)
```

```python
plot_hist_features(metrics_lr, metrics_rf, "unbalanced_acc")
plot_hist_features(metrics_lr, metrics_rf, "unbalanced_prec")
plot_hist_features(metrics_lr, metrics_rf, "unbalanced_rec")
plot_hist_features(metrics_lr, metrics_rf, "unbalanced_f1")
```

```python
plot_hist_features(metrics_lr, metrics_rf, "balanced_acc")
plot_hist_features(metrics_lr, metrics_rf, "balanced_prec")
plot_hist_features(metrics_lr, metrics_rf, "balanced_rec")
plot_hist_features(metrics_lr, metrics_rf, "balanced_f1")
```

```python
plot_hist_features(metrics_lr, metrics_rf, "balanced_zeros_acc")
plot_hist_features(metrics_lr, metrics_rf, "balanced_zeros_prec")
plot_hist_features(metrics_lr, metrics_rf, "balanced_zeros_rec")
plot_hist_features(metrics_lr, metrics_rf, "balanced_zeros_f1")
```
