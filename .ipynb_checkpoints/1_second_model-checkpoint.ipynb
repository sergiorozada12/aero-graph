# 2.1) New variables about particular delays

```python
N = nodes.shape[0]
E = edges.shape[0]
hours = df.shape[0]/N
```

```python
# INCOMING MEAN DELAY
df_work = df[['NODE', 'MEAN_DELAY']].copy()

for node in nodes:
    df_work[node + '_MEAN_DELAY'] = np.repeat(df_work.loc[df_work['NODE'] == node, 'MEAN_DELAY'].values, N)

neighbors = {node: [neigh + '_MEAN_DELAY' for neigh in edges[np.where(edges[:, 1] == node), 0][0]] for node in nodes}
incoming_mean_delay = df_work.apply(lambda row: row[neighbors[row['NODE']]].sum(), axis=1).values
del df_work

print("DONE")
```

```python
# INCOMING MEAN CARRIER DELAY
df_work = df[['NODE', 'MEAN_CARRIER_DELAY']].copy()

for node in nodes:
    df_work[node + '_MEAN_CARRIER_DELAY'] = np.repeat(df_work.loc[df_work['NODE'] == node, 'MEAN_CARRIER_DELAY'].values, N)

neighbors = {node: [neigh + '_MEAN_CARRIER_DELAY' for neigh in edges[np.where(edges[:, 1] == node), 0][0]] for node in nodes}
incoming_mean_carrier_delay = df_work.apply(lambda row: row[neighbors[row['NODE']]].sum(), axis=1).values
del df_work

print("DONE")
```

```python
# INCOMING MEAN LATE AIRCRAFT DELAY
df_work = df[['NODE', 'MEAN_LATE_AIRCRAFT_DELAY']].copy()

for node in nodes:
    df_work[node + '_MEAN_LATE_AIRCRAFT_DELAY'] = np.repeat(df_work.loc[df_work['NODE'] == node, 'MEAN_LATE_AIRCRAFT_DELAY'].values, N)

neighbors = {node: [neigh + '_MEAN_LATE_AIRCRAFT_DELAY' for neigh in edges[np.where(edges[:, 1] == node), 0][0]] for node in nodes}
incoming_mean_late_aircraft_delay = df_work.apply(lambda row: row[neighbors[row['NODE']]].sum(), axis=1).values
del df_work

print("DONE")
```

```python
# INCOMING MEAN NAS DELAY
df_work = df[['NODE', 'MEAN_NAS_DELAY']].copy()

for node in nodes:
    df_work[node + '_MEAN_NAS_DELAY'] = np.repeat(df_work.loc[df_work['NODE'] == node, 'MEAN_NAS_DELAY'].values, N)

neighbors = {node: [neigh + '_MEAN_NAS_DELAY' for neigh in edges[np.where(edges[:, 1] == node), 0][0]] for node in nodes}
incoming_mean_nas_delay = df_work.apply(lambda row: row[neighbors[row['NODE']]].sum(), axis=1).values
del df_work

print("DONE")
```

```python
# INCOMING MEAN SECURITY DELAY
df_work = df[['NODE', 'MEAN_SECURITY_DELAY']].copy()

for node in nodes:
    df_work[node + '_MEAN_SECURITY_DELAY'] = np.repeat(df_work.loc[df_work['NODE'] == node, 'MEAN_SECURITY_DELAY'].values, N)

neighbors = {node: [neigh + '_MEAN_SECURITY_DELAY' for neigh in edges[np.where(edges[:, 1] == node), 0][0]] for node in nodes}
incoming_mean_security_delay = df_work.apply(lambda row: row[neighbors[row['NODE']]].sum(), axis=1).values
del df_work

print("DONE")
```

```python
# INCOMING MEAN WHEATHER DELAY
df_work = df[['NODE', 'MEAN_WEATHER_DELAY']].copy()

for node in nodes:
    df_work[node + '_MEAN_WEATHER_DELAY'] = np.repeat(df_work.loc[df_work['NODE'] == node, 'MEAN_WEATHER_DELAY'].values, N)

neighbors = {node: [neigh + '_MEAN_WEATHER_DELAY' for neigh in edges[np.where(edges[:, 1] == node), 0][0]] for node in nodes}
incoming_mean_weather_delay = df_work.apply(lambda row: row[neighbors[row['NODE']]].sum(), axis=1).values
del df_work

print("DONE")
```

```python
df['INCOMING_DELAY'] = incoming_mean_delay
df['INCOMING_CARRIER_DELAY'] = incoming_mean_carrier_delay
df['INCOMING_LATE_AIRCRAFT_DELAY'] = incoming_mean_late_aircraft_delay
df['INCOMING_NAS_DELAY'] = incoming_mean_nas_delay
df['INCOMING_SECURITY_DELAY'] = incoming_mean_security_delay
df['INCOMING_WEATHER_DELAY'] = incoming_mean_weather_delay
```

# 5) Median Modelling

```python
# COLUMNS
cols_time = ['HOUR', 'DAY', 'DAY_OF_WEEK', 'MONTH', 'QUARTER', 'YEAR']                 # time variables
cols_indicator = [col for col in df.columns if 'NODE_' in col]                         # reference node
cols_delay = [col for col in df.columns if '_DELAY' in col and 'MEDIAN_' not in col]   # all delays
col_local_delay = ['MEDIAN_DELAY']                                                     # local delay
col_incoming_delay = ['INCOMING_DELAYS']                                               # incoming delays
col_target = ['y']                                                                     # target
```

```python
# TRAIN-TEST SPLIT
train_size = .8

#idx_train = np.random.choice(df.index, int(train_size*df.shape[0]), replace=False)
#idx_test = np.setdiff1d( df.index, idx_train)
idx_train = np.where(df['YEAR'] == 2018)
idx_test = np.where(df['YEAR'] == 2019)

df_train = df.iloc[idx_train[0], :].reset_index(drop=True)
df_test = df.iloc[idx_test[0], :].reset_index(drop=True)
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
print("Number of delays: ", (df_train['y'] == 1).sum())
```

```python
# UNDERSAMPLED DF
ratio_imbalance = 3

random_indices = np.random.choice(np.where(df_train[col_target] == 0)[0], 
                                  ratio_imbalance*np.sum(df_train[col_target] == 1).iloc[0], 
                                  replace=False)

idx = np.concatenate([random_indices, np.where(df_train[col_target] == 1)[0]])

df_undersampled = df_train.loc[idx, :].reset_index(drop=True)
```

```python
print(df_undersampled.shape)
```

### 5.1) First dataset: Airport information

```python
X_train = df_undersampled[cols_time + cols_indicator + col_local_delay].values
y_train = df_undersampled[col_target].values

X_test = df_test[cols_time + cols_indicator + col_local_delay].values
y_test = df_test[col_target].values

scaler = MinMaxScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42, shuffle=True)
```

```python
model = Sequential()
model.add(Dense(1000, activation='relu', input_dim=X_train.shape[1]))
model.add(Dropout(0.5))
model.add(Dense(1000, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(y_train.shape[1], activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

es = EarlyStopping(monitor='val_loss', 
                   mode='min', 
                   verbose=1, 
                   patience=10)

history = model.fit(X_train, 
                    y_train,
                    validation_data=(X_val, y_val),
                    epochs=100,
                    batch_size=128,
                    callbacks=[es],
                    shuffle=True)
```

```python
fig, ax = plt.subplots(1, 2, figsize=[18, 4])
ax = ax.flatten()

ax[0].plot(history.history['loss'], label='train')
ax[0].plot(history.history['val_loss'], label='test')
ax[0].legend()

ax[1].plot(history.history['acc'], label='train')
ax[1].plot(history.history['val_acc'], label='test')
ax[1].legend()
plt.show()
```

```python
y_pred = (model.predict(X_test) >= .5)*1

acc_1 = accuracy_score(y_pred, y_test)
prec_1 = precision_score(y_pred, y_test)
rec_1 = recall_score(y_pred, y_test)

print("Accuracy: ", acc_1)
print("Precision: ", prec_1)
print("Recall: ", rec_1)
```

```python
acc_1 = 0.9315086731588558
prec_1 = 0.1132657558395769
rec_1 = 0.13921993499458288
```

### 5.4) LSTM

```python
del df, idx_train, idx_test

n_steps = 8
cols_train = cols_time + cols_indicator + cols_delay + col_incoming_delay

n_cols_X = len(cols_train)
n_cols_y = len(col_target)

n_rows = df_train.shape[0]//1

X_train = df_train.head(n_rows)[cols_train].values
y_train = df_train.head(n_rows)[col_target].values

del df_train

X_test = df_test[cols_train].values
y_test = df_test[col_target].values

del df_test
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train = X_train.reshape(-1, n_steps, n_cols_X)
X_test = X_test.reshape(-1, n_steps, n_cols_X)

y_train = np.array([arr[-1] for arr in y_train.reshape(-1, n_steps, n_cols_y)])
y_test = np.array([arr[-1] for arr in y_test.reshape(-1, n_steps, n_cols_y)])
```

```python
# UNDERSAMPLED DF FOR LSTM
ratio_imbalance = 3

random_indices = np.random.choice(np.where(y_train == 0)[0],
                                  ratio_imbalance*np.sum(y_train == 1), 
                                  replace=False)

idx = np.concatenate([random_indices, np.where(y_train == 1)[0]])

X_train = X_train[idx, :]
y_train = y_train[idx]
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42, shuffle=True)
```

```python
model = Sequential()
model.add(LSTM(1000, input_shape=(n_steps, n_cols_X)))
model.add(Dropout(0.5))
model.add(Dense(1000, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1000, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(y_train.shape[1], activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

es = EarlyStopping(monitor='val_loss', 
                   mode='min', 
                   verbose=1, 
                   patience=10)

history = model.fit(X_train, 
                    y_train,
                    validation_data=(X_val, y_val),
                    epochs=100,
                    batch_size=128,
                    callbacks=[es],
                    shuffle=True)
```

```python
fig, ax = plt.subplots(1, 2, figsize=[18, 4])
ax = ax.flatten()

ax[0].plot(history.history['loss'], label='train')
ax[0].plot(history.history['val_loss'], label='test')
ax[0].legend()

ax[1].plot(history.history['acc'], label='train')
ax[1].plot(history.history['val_acc'], label='test')
ax[1].legend()
plt.show()
```

```python
y_pred = (model.predict(X_test) >= .5)*1

acc_4 = accuracy_score(y_pred, y_test)
prec_4 = precision_score(y_pred, y_test)
rec_4 = recall_score(y_pred, y_test)

print("Accuracy: ", acc_4)
print("Precision: ", prec_4)
print("Recall: ", rec_4)
```
