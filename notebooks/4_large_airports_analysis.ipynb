```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
DATA_PATH = "data/"
```

```python
imp_nodes = np.loadtxt(DATA_PATH + "biggest_airports.txt", dtype=str)
```

```python
df = pd.read_csv(DATA_PATH + "incoming_delays_nodes.csv", sep='|')[["NODE", "y_clas"]]
```

```python
nodes_count = df[df["y_clas"] == 1]["NODE"].value_counts()
```

```python
color = ['r' if node in imp_nodes else 'b' for node in nodes_count.index]
```

```python
#9.5K horas totales
plt.figure(figsize=[18, 4])
plt.title("Top 100 retrasos")
plt.bar(x=nodes_count.index[:100], height=nodes_count.values[:100], color=color[:100])
plt.xticks(rotation=90)
plt.legend(['important'])
plt.show()
```

```python
#9.5K horas totales
plt.figure(figsize=[18, 4])
plt.title("Distribuci√≥n de los retrasos")
plt.bar(x=np.arange(nodes_count.size), height=nodes_count.values, color=color)
plt.xticks(rotation=90)
plt.legend(['important'])
plt.show()
```
