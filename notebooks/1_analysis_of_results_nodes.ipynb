```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
DATA_PATH = "C:\\Users\\victor\\Documents\\Aero_TFG\\results\\results0317\\"
```

```python
MIN_ACC_PAPER = 0.712
MAX_ACC_PAPER = 0.887
MEAN_ACC_PAPER = 0.809
```

```python
def get_data(results, zeros=True):
    unbalanced_acc = []
    balanced_acc = []
    balanced_zeros_acc = []

    unbalanced_prec = []
    balanced_prec = []
    balanced_zeros_prec = []

    unbalanced_rec = []
    balanced_rec = []
    balanced_zeros_rec = []
    
    unbalanced_f1 = []
    balanced_f1 = []
    balanced_zeros_f1 = []
    
    for node in results.keys():
        unbalanced_acc.append(results[node]['unbal']['acc'])
        balanced_acc.append(results[node]['bal']['acc'])

        unbalanced_prec.append(results[node]['unbal']['prec'])
        balanced_prec.append(results[node]['bal']['prec'])

        unbalanced_rec.append(results[node]['unbal']['rec'])
        balanced_rec.append(results[node]['bal']['rec'])
        
        unbalanced_f1.append(results[node]['unbal']['f1'])
        balanced_f1.append(results[node]['bal']['f1'])
        
        if zeros:
            balanced_zeros_acc.append(results[node]['bal_zeros']['acc'])
            balanced_zeros_prec.append(results[node]['bal_zeros']['prec'])
            balanced_zeros_rec.append(results[node]['bal_zeros']['rec'])
            balanced_zeros_f1.append(results[node]['bal_zeros']['f1'])
    
    return {'unbalanced_acc': unbalanced_acc,
            'unbalanced_prec': unbalanced_prec,
            'unbalanced_rec': unbalanced_rec,
            'unbalanced_f1': unbalanced_f1,
            'balanced_acc': balanced_acc,
            'balanced_prec': balanced_prec,
            'balanced_rec': balanced_rec,
            'balanced_f1': balanced_f1,
            'balanced_zeros_acc': balanced_zeros_acc,
            'balanced_zeros_prec': balanced_zeros_prec,
            'balanced_zeros_rec': balanced_zeros_rec,
            'balanced_zeros_f1': balanced_zeros_f1}
```

```python
def plot_hist_features(metrics_1, metrics_2, exp_type):
    plt.figure(figsize=[16, 3])
    plt.hist(metrics_1[exp_type], range=[0, 1], bins=50, color='b', alpha=.5)
    if metrics_2 != None:
        plt.hist(metrics_2[exp_type], range=[0, 1], bins=50, color='r', alpha=.5)
    if exp_type == 'balanced_acc':
        plt.axvline(x=MIN_ACC_PAPER, color='c')
        plt.axvline(x=MAX_ACC_PAPER, color='m')
        plt.axvline(x=MEAN_ACC_PAPER, color='g')
    plt.xlabel(exp_type)
    plt.legend(['LR', 'RF'])
    plt.show()
```

```python
def plot_bar(metrics_1, metrics_2, labels, title):
    x = np.arange(len(labels))
    
    plt.figure(figsize=[16, 3])
    plt.title(title)
    plt.bar(x=x-.25, height=metrics_1, color='b', width=.25)
    plt.bar(x=x, height=metrics_2, color='g', width=.25)
    plt.xticks(x, labels)
    plt.legend(['LR', 'RF'])
    plt.show()
```
